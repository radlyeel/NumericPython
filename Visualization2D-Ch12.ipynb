# Basic plot, Fig. 12.1

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
x = [0, 1, 2, 3]
y = [0, 1, 4, 9]
plt.plot(x, y)
plt.show()
```

Make a plot of the function f(x) = x^2 for âˆ’5 â‰¤ x â‰¤ 5 with dashed green line(error in Fig 12.3)

```python
%matplotlib inline
x = np.linspace(-5,5, 100)
plt.plot(x, x**2, "g--")
plt.show()
```

plot two functions

```python
x = np.linspace(-5,5,20)
plt.plot(x, x**2, "ko")
plt.plot(x, x**3, "r*")
plt.show()
```

Decorate the plot

```python
plt.figure(figsize = (10,6))
x = np.linspace(-5,5,20)
plt.plot(x, x**2, "ko")
plt.plot(x, x**3, "r*")
plt.title(f"Plot of x**2 and x**3 from {x[0]} to{ x[-1]}")

plt.xlabel("X axis", fontsize = 18)
plt.ylabel("Y axis", fontsize = 18)
plt.show()
```

List available styles


```python
print(plt.style.available)
```

use one of them, and add legend

```python
plt.style.use("seaborn-poster")
plt.figure(figsize = (10,6))
x = np.linspace(-5,5,20)
plt.plot(x, x**2, "ko", label = "quadratic")
plt.plot(x, x**3, "r*", label = "cubic")
plt.title(f"Plot of x**2 and x**3 from {x[0]} to{x[-1]}")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend(loc = 2)
plt.show()
```

Set limits and add grid

```python
plt.figure(figsize = (10,6))
x = np.linspace(-5,5,100)
plt.plot(x, x**2, "ko", label = "quadratic")
plt.plot(x, x**3, "r*", label = "cubic")
plt.title(f"Plot of x**2 and x**3 from {x[0]} to {x[-1]}")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend(loc = 2)
plt.xlim(-6.6)
plt.ylim(-10,10)
plt.grid()
plt.show()
```

Multiple plots in same figure, and save as PDF

```python
x = np.arange(11)
y = x**2
plt.figure(figsize = (14, 8))
plt.subplot(2, 3, 1)
plt.plot(x,y)
plt.title("Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.subplot(2, 3, 2)
plt.scatter(x,y)
plt.title("Scatter")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.subplot(2, 3, 3)
plt.bar(x,y)
plt.title("Bar")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.subplot(2, 3, 4)
plt.loglog(x,y)
plt.title("Loglog")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(which="both")
plt.subplot(2, 3, 5)
plt.semilogx(x,y)
plt.title("Semilogx")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(which="both")
plt.subplot(2, 3, 6)
plt.semilogy(x,y)
plt.title("Semilogy")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.tight_layout()
#plt.show()
plt.savefig("image.pdf")
```
