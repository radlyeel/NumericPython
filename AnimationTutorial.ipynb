```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib notebook
```

```python
 xdata = []
 ydata = []
 (fig, ax) = plt.subplots()
 ax.set_xlim(0,105)
 ax.set_ylim(0,12)
 line, = ax.plot(0,0)
def animation_fn(i):
     xdata.append(i * 10)
     ydata.append(i)
     line.set_xdata(xdata)
     line.set_ydata(ydata)
     return line,
animation = FuncAnimation(fig, animation_fn, frames=np.arange(0, 10, 0.01), interval=10)
plt.show()
```
